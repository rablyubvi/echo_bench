# Sinh dữ liệu

Viết một hàm có thể trích xuất phần văn bản là lời tự sự - dữ liệu thực sự cần thiết

```python
def extract_narrative_expression(text):
    # Sử dụng regex để tìm các phần tử sau số thứ tự
    matches = re.findall(r'\d+\.\s(.+)', text)
    matches = [match.rstrip('.') for match in matches]
    return matches
```

Sinh dữ liệu với MOOD là 'Happy'

```python
%%time
response = model.generate_content("Sinh ra 150 lời tâm sự dạng nhật ký có thể có đại từ hoặc không, diễn tả 1 sự việc xảy ra hoặc 1 trạng thái của người đó mà trong đó diễn tả niềm vui hoặc hạnh phúc, với văn phong khác nhau")
print(response.text)
```

```python
text = response.text

extracted_texts = extract_narrative_expression(text)
for line in extracted_texts:
    print(line)
```

```python
happy_data = {'COMMENT': extracted_texts}
happy_df = pd.DataFrame(happy_data)
happy_df['MOOD'] = 'Happy'
```
